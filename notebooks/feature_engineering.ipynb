# Preprocessing and Feature engineering

```python
import numpy as np
import pandas as pd
```

```python
# Reading csv
df = pd.read_csv('../data/Telco-Customer-Churn.csv')

# Dropping customer_id
df.drop(columns=['customerID'], inplace=True)

# Keeping Phone Service, Multiple Lines, Streaming cols despite low visual signal.
# Will revisit after SHAP importance analysis in Week 3.
```

##### Encoding

```python
# Encdoing Churn 
df['Churn'] = (df['Churn'] == 'Yes').astype(int)

# Encoding Total Charges as numeric and filling missing values with zero
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Encoding Categoricals
df = pd.get_dummies( df, columns = ['Contract', 'PaymentMethod', 'InternetService'], dtype=int )
```

```python
binary_cols = df.select_dtypes(include='object').columns

# Encode 
df[binary_cols] = df[binary_cols].apply(lambda col: col.map(
    {'Yes': 1, 'No': 0, 'Female': 1, 'Male': 0,
     'No phone service': 0, 'No internet service': 0}
))

df[binary_cols].isnull().sum()
```
